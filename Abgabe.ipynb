### c)

Das zweite Optimierungsproblem ist wohldefiniert, da der natürliche Logarithmus streng monoton steigend auf dem gesamten Definitionsbereich ist.
Für $f(x) = ln(x)$ gilt $D \textsubscript{f} = R\textsuperscript{+}$, also der Definitionsbereich des natürlichen Logarithmus sind die positiven, reellen Zahlen, ausgenommen der Null. Aus der ersten Restriktion ($min(x,y) \geq 1$) und $ln(z)$ mit $z=x+2y$ folgt $z \geq 3$. Aus dieser Erkenntnis lässt sich zusammen mit dem streng monotonen Wachstum folgern, dass das Optimierungsproblem wohldefiniert ist, da sich eindeutig ein Maximum oder Minimum bestimmen lässt.

## Lösung der Aufgabe 3


### a)

Damit das zweidimensionale LP einen eindeutigen Optimalpunkt besitzt, wurden die folgenden Nebenbedingungen definiert:

1. $x_{1}\le -1 $
2. $x_{2}\le -1 $
3. $x_{1} * x_{2} \ge 1$

Aus der definierten Polyedrischen Menge M folgt, dass der Optimalpunkt bei $(x_{1} =-1 , x_{2} =-1)$ liegt. Daraus folgt der Optimalwert von -2.


### b)

Die Menge M wurde zudem so konstruiert, dass eine beliebige Restriktion entfernt werden kann und sich der Optimalpunkt trotzdem nicht ändert:

- Streichen der ersten Nebenbedingung:
    Aus der Bedingung $x_{2}\le -1 $ folgt, dass der ideale Wert für $x_{2}$ bei $-1 $ liegt. Aus der Erkenntnis und der Bedingung $x_{1} * x_{2} \ge 1$ lässt sich dann folgern, dass auch $x_{1}$ idealerweise bei $-1 $ liegen sollte.

- Streichen der zweiten Nebenbedingung:
    Hier lässt sich anlog dem Vorgehen von oben aus der Bedingung $x_{1}\le -1 $ entnehmen, dass der ideale Wert für $x_{1}$ idealerweise bei $-1 $ liegt. Aus der Erkenntnis und der Bedingung $x_{1} * x_{2} \ge 1$ lässt sich dann folgern, dass auch $x_{2}$ idealerweise wieder bei $-1 $ liegen sollte.

- Streichen der dritten Nebenbedingung:
    Hier lässt sicher wiederum direkt ablesen, dass auf Grund der Bedingung $x_{1}\le -1 $ der ideale Wert für $x_{1}$ bei $-1$ liegt. Die Bedingung $x_{2}\le -1 $ führt dann wiederum dazu, dass der Wert für $x_{2}$ auch wieder idealweise bei $-1$ liegen sollte.

Egal welche der drei Bedingungen weggelassen wird, der Optimalpunkt liegt weiterhin bei $(x_{1} =-1 , x_{2} =-1)$.

## Lösung der Aufgabe 4:

### a)




Um das Problem lösen zu können, müssen sowohl die Entscheidungsvariablen, die Constraints als auch der Optimierungsterm festgelegt werden.

**Entscheidungsvariablen**

Wie im Unttericht demonstriert wurde, ist das Lösen solcher Probleme mit dem One-Hot-Encoding deutlich effizienter möglich, als mit einem Linearen Wert in der entsprechenden Zelle. Daher wurde sich dafür entschieden, neben der Zeile und der Spalte auch noch für die Farbe einen eigenen Index zu definieren.  
Dabei sind die Indizes bei den Farben für folgende Farben definiert:
- 1: gelb  
- 2: lila  
- 3: rot  
- 4: grün  
- 5: schwarz  
  
Die Entscheidungsvariable lautet somit:
$$    x_{r,c,f} $$
mit c $\in$ C = {1,2....,10}, r $\in$ R = {1,2,..,5} und f $\in$ F = {1,2,3,4,5}

Um das Problem entsprechend seinen Vorgaben zu modellieren, wurden die folgenden Constraints aufgestellet:

1) Startpunkte müssen eingehalten werden::
$$x_{r,c,f}*f = w_{r,c}, r\in R, c\in C $$  
  
2) An jeder Position darf nur eine Farbe=1 sein:  
$$\sum_{f \in F} x_{r,c,f} = 1, r\in R, c\in C$$

3) Von jeder Farbe muss die gleiche Anzahl(10) auf dem Spielfeld sein:  
$$\sum_{r \in R}\sum_{c \in c} x_{r,c,f} = \frac{r*c}{f}, f \in F$$

4) Wenn eine Farbe auf einer Position ist, muss links, rechts, oben oder unten mindestens ein weitere Position mit der selben Farbe sein  
$$wenn\: x_{r,c,f} ==1 : x_{r-1,c,f}+x_{r-1,c,f}+x_{r,c-1,f}+x_{r,c+1,f} \geq 1, c \in C, r \in R, f \in F, $$

5) Alle Zellen einer Farbe müssen miteinander verbunden sein:  
  
6) Es dürfen keine Rechtecke entstehen:  
$$x_{r,c,f}+x_{r-1,c,f}+x_{r-1,c-1,f}+x_{r,c-1,f} \leq 3, c \in C, r \in R, f \in F, $$

```python
import time
from mip import CBC, GUROBI, Model, BINARY, xsum
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
solver = GUROBI
```

Funktion zum Visualisieren der Graphen

```python
def visualize(matrix):
    color_map = []
    for n in np.nditer(matrix):
        if n == 0:
            color_map.append('orange')
        if n == 1:
            color_map.append('purple')
        if n == 2:
            color_map.append('red')
        if n == 3:
            color_map.append('green')
        if n == 4:
            color_map.append('black')


    G = nx.grid_2d_graph(10, 5)

    plt.figure(figsize=(4, 8))
    pos = {(x,y):(y,-x) for x,y in G.nodes()}
    nx.draw(
        G, 
        pos, 
        node_color = color_map, 
        with_labels = False,
        node_size = 500,
    )
```

```python
def getNeighbors(row, column):
    neighbor_rows = []
    neighbor_columns = []

    if(row-1 >= 0):
        neighbor_rows.append(row-1)
    if(row+1 <= 9):
        neighbor_rows.append(row+1)
    if(column-1 >= 0):
        neighbor_columns.append(column-1)
    if(column+1 <= 4):
        neighbor_columns.append(column+1)

    neighbors = []

    for r in neighbor_rows:
        neighbors.append([r, column])

    for c in neighbor_columns:
        neighbors.append([row, c])

    return neighbors
```

```python
def getSquare(row, column):
    squares = []

    for r in [row, row-1]:
        for c in [column, column-1]:
            squares.append([r,c])

    return squares
```

```python
# track Execution time
t_start=time.time()

# Data Structure
rows = range(0, 10)         # Anzahl der Reihen, nummeriert von 1 bis 10
columns = range(0, 5)       # Anzahl der Spalten
values = range(0, 5)        # Anzahl der möglichen Farben

#Model declaration

m = Model("Worms", solver_name=solver)
    
# Decision Variables
y = {(r,c,v): m.add_var(var_type=BINARY) for r in rows for c in columns for v in values}

# Constraints
## Respect inital values
m += y[7,2,0] == 1 #Gelb(0) bei 22 
m += y[6,4,1] == 1 #Lila(1) bei 43
m += y[2,2,2] == 1 #Rot(2) bei 27
m += y[7,1,3] == 1 #Grün(3) bei 12
m += y[5,1,4] == 1 #Schwarz(4) bei 14

## Only one color at each point
for r in rows:
    for c in columns:
        m += xsum(y[r,c,v] for v in values) == 1

## Same number of points per color
for v in values:
    m += xsum(y[r,c,v] for r in rows for c in columns) == 10

## at least one neighbor with same color
for v in values:
    for r in rows:
        for c in columns:
            m += xsum(y[a,b,v] for [a,b] in getNeighbors(r,c)) >= y[r,c,v]
            

# Optimize Statement
m.optimize()

# Track and print execution time
t_end = time.time()
print(f" Elapsed second for solution with {solver}: {t_end-t_start} seconds." )


color_map = np.zeros((10,5))

for r in rows:
    for c in columns:
        for v in values:
            if y[r,c,v].x == 1.0:
                color_map[r,c] = v

visualize(color_map)
```
