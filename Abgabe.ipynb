## Lösung der Aufgabe 3


### a)

Damit das zweidimensionale LP einen eindeutigen Optimalpunkt besitzt, wurden die folgenden Nebenbedingungen definiert:

1. $x_{1}\le -1 $
2. $x_{2}\le -1 $
3. $x_{1} * x_{2} \ge 1$

Aus der definierten Polyedrischen Menge M folgt, dass der Optimalpunkt bei $(x_{1} =-1 , x_{2} =-1)$ liegt. Daraus folgt der Optimalwert von -2.


### b)

Die Menge M wurde zudem konstruiert, dass eine beliebige Restriktion entfernt werden kann und sich der Optimalpunkt trptzdem nicht ändert:

- Streichen der ersten Nebenbedingung:  
    Aus der Bedigung $x_{2}\le -1 $ folgt, dass der ideale Werte für $x_{2}$ idealerweise bei $-1 $ liegt. Aus der Erkenntnis und der Bedingung $x_{1} * x_{2} \ge 1$ lässt sich dann folgern, dass auch $x_{1}$ idealerweise bei $-1 $ liegen sollte. 
      
- Streichen der zweiten Nebenbedigung:
    Hier lässt sich anlog dem Vorgehen von oben aus der Bedigung $x_{1}\le -1 $ entnehmen, dass der ideale Werte für $x_{1}$ idealerweise bei $-1 $ liegt. Aus der Erkenntnis und der Bedingung $x_{1} * x_{2} \ge 1$ lässt sich dann folgern, dass auch $x_{2}$ idealerweise wieder bei $-1 $ liegen sollte. 
      
- Streichen der dritten Nebenbedigung:
    Hier lässt sicher wiedrum direkt ablesen, dass auf Grund der Bedingung $x_{1}\le -1 $ der ideale Wert für $x_{1}$ bei $-1$ liegt. Die Bedingung $x_{2}\le -1 $ führt dann wiederum dazu, dass der Wert für $x_{2}$ auch wieder idealweise bei $-1$ liegen sollte.
      
Egal welche der drei Bedingungen weggelassen wird, der Optimalpunkt liegt weiterhin bei $(x_{1} =-1 , x_{2} =-1)$.

Wie im Unttericht demonstriert wurde, ist das Lösen solcher Probleme mit dem One-Hot-Encoding deutlich effizienter möglich, als mit einem Linearen Wert in der entsprechenden Zelle. Daher wurde sich dafür entschieden, neben der Zeile und der Spalte auch noch für die Farbe einen eigenen Index zu definieren.  
Dabei sind die Indizes bei den Farben für folgende Farben definiert:
- 1: gelb  
- 2: lila  
- 3: rot  
- 4: grün  
- 5: schwarz  
  
Die Entscheidungsvariable lautet somit:
$$    x_{r,c,f} $$
mit c $\in$ C = {1,2....,10}, r $\in$ R = {1,2,..,5} und f $\in$ F = {1,2,3,4,5}

Um das Problem entsprechend seinen Vorgaben zu modellieren, wurden die folgenden Constraints aufgestellet:

1) Startpunkte müssen eingehalten werden::
$$x_{r,c,f}*f = w_{r,c}, r\in R, c\in C $$  
  
2) An jeder Position darf nur eine Farbe=1 sein:  
$$\sum_{f \in F} x_{r,c,f} = 1, r\in R, c\in C$$

3) Von jeder Farbe muss die gleiche Anzahl(10) auf dem Spielfeld sein:  
$$\sum_{r \in R}\sum_{c \in c} x_{r,c,f} = \frac{r*c}{f}, f \in F$$

4) Wenn eine Farbe auf einer Position ist, muss links, rechts, oben oder unten mindestens ein weitere Position mit der selben Farbe sein  
$$wenn\: x_{r,c,f} ==1 : x_{r-1,c,f}+x_{r-1,c,f}+x_{r,c-1,f}+x_{r,c+1,f} \geq 1, c \in C, r \in R, f \in F, $$

5) Alle Zellen einer Farbe müssen miteinander verbunden sein:  
  
6) Es dürfen keine Rechtecke entstehen:  
$$x_{r,c,f}+x_{r-1,c,f}+x_{r-1,c-1,f}+x_{r,c-1,f} \leq 3, c \in C, r \in R, f \in F, $$
